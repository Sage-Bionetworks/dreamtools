# D4C3 challenge ( Predictive Signaling Network Modeling)

This challenge explores the extent to which our current knowledge of signaling pathways, collected from a variety of cell types, agrees with cell-type specific high-throughput experimental data. 

See https://www.synapse.org/#!Synapse:syn2825304/wiki/ for details

```python
%pylab inline
```

```python
from dreamtools import D4C3
```

```python
s = D4C3()
```

```python
# Retrieves a path to a template
filename = s.download_template()
```

```python
results = s.score(filename)
```

```python
# The returned object is a Pandas object with metrics about performance
# of that submissions
results
```

```python
# Note that this challenge has a plot method, which is not provided
# in all challenges
s.plot()
```

```python
# You can also download the gold standard and check its performance...
```

```python
gs = s.download_goldstandard()
```

```python
results = s.score(gs)
```
